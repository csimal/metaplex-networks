```julia
using Plots
using Graphs
using NetworkEpidemics
using MATLAB
```

```julia
include("fig_utils.jl")
```

```julia
N = 1000
M = 10
```

```julia
x0_i = fill(1, N)
x0_i[1:10] .= 2
x0_μ = [div(i-1,div(N,M))+1 for i in 1:N];
```

```julia
x0_mp = Array{Int,2}(undef, M, 2)
x0_mp[2:M,:] .= [100 0]
x0_mp[1, :] .= [90, 10];
```

```julia
g = erdos_renyi(N, 0.3)
h = complete_graph(M)

fig_er = CorrectedMetapopulationFigure(g, h, SI(0.1), 0.1, x0_i, x0_μ, x0_mp);
```

```julia
compute_figure_data!(fig_er)
```

```julia
make_plot_julia(fig_er)
```
